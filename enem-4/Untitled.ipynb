```python
import pandas as pd
from sklearn.ensemble import RandomForestRegressor
```

```python
train = pd.read_csv('train.csv')
test = pd.read_csv('test.csv')
```

```python
print(train.shape)
print(test.shape)
```

```python
bool_cols = train.columns.isin(test.columns)
```

```python
df_train = pd.concat([train[train.columns[bool_cols]], train['IN_TREINEIRO']], axis = 1)
```

```python
df_aux = pd.DataFrame({
                        'tipo': df_train.dtypes,
                        'null' : df_train.isna().sum(),
                            'colunas' : df_train.shape[0],
                        '%'    : 100*df_train.isna().sum()/df_train.shape[0],
                        'qtde valores' : df_train.nunique(),
})
```

```python
df_aux
```

```python
correlation = df_train.corr()['IN_TREINEIRO']
```

```python
col_train = []
for l in range(len(correlation)):
    if correlation[l] >.2 or correlation[l] < -0.2:
        col_train.append(correlation.index[l])


col_test = col_train[0:-1]
```

```python
pd.DataFrame({
            'tipos': train[col_train].dtypes,
            'vazios': train[col_train].isna().sum()
})
```

```python
train_ = train[col_train]
```

```python
test_ = test[col_test]
```

```python
train_x = train_.drop('IN_TREINEIRO', axis=1)
```

```python
train_x
```

```python
train_y = train_['IN_TREINEIRO']
```

```python
test_
```

```python
ranforest = RandomForestRegressor().fit(train_x, train_y)
```

```python
test_y = ranforest.predict(test_)
```

```python
test_y
```

```python
answer = pd.DataFrame({
                        'NU_INSCRICAO': test.NU_INSCRICAO,
                        'IN_TREINEIRO' : test_y
})
```

```python
answer.to_csv('answer.csv', index=False)
```

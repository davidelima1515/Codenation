```python
import pandas as pd
import numpy as np
import seaborn as sns 
from sklearn.datasets import load_boston
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error
from math import sqrt
from sklearn.preprocessing import OneHotEncoder
```

```python
carregando_df = load_boston()
```

```python
carregando_df.feature_names
```

```python
print(carregando_df.DESCR)
```

```python
df = pd.DataFrame(carregando_df.data, columns=carregando_df.feature_names)
df.head(8)
```

```python
df_target = pd.DataFrame(data = carregando_df.target, columns=["Target"])
df_target
```

```python
df_concat = pd.concat([df, df_target],axis=1,ignore_index=False)
```

```python
df_aux = pd.DataFrame({
                        "VAZIOS" : df_concat.isna().sum(),
                        "% VAZIOS": df_concat.isna().sum().mean(), 
                        "TIPOS": df_concat.dtypes,
                        "VALORES ÚNICOS": df_concat.nunique(),
                        "MÉDIA": df_concat.mean()
})

df_aux
```

```python
df_concat.corr()
```

```python
data = df_concat.corr()
```

```python
plt.figure(figsize=(10,10))
sns.heatmap(data[(data>0.4) | (data<-.4)], annot= True)
```

```python
x_train, x_test, y_train, y_test = train_test_split(df, df_target, test_size = .2, random_state = 80)
```

```python
reg = LinearRegression().fit(x_train, y_train)
```

```python
target_pred = reg.predict(x_test)
```

```python
mean_squared_error(y_test, target_pred)
```

```python
r2_score(y_test, target_pred)
```

```python
mean_absolute_error(y_test, target_pred)
```
